# 01 - Deep Agents 概述

**来源:** [LangChain Docs - Deep Agents Overview](https://docs.langchain.com/oss/python/deepagents/overview)

Deep Agents 是一个 **Agent 框架(Harness)**，在 LangChain 核心组件和 LangGraph 运行时之上构建。

它内置了规划、文件系统、子 Agent、长期记忆等开箱即用的能力。

## 1. 一句话创建 Agent

Deep Agents 用最少的样板代码就能创建一个可用 Agent：

````python
# 安装: pip install -qU deepagents

from deepagents import create_deep_agent

def get_weather(city: str) -> str:
    return f"It's always sunny in {city}!"

agent = create_deep_agent(
    model="deepseek-v4-flash",
    tools=[get_weather],
    system_prompt="You are a helpful assistant",
)

result = agent.invoke({
    "messages": [{"role": "user", "content": "what is the weather in sf"}]
})
print(result["messages"][-1].content)
````

## 3. 核心能力(Harness Capabilities)

| 组件 | 包含能力 |
|------|----------|
| 执行环境 | 工具、虚拟文件系统、可选沙箱、解释器(REPL) |
| 上下文管理 | Skills、Memory、摘要、上下文卸载、Prompt Caching |
| 委派 | 子 Agent 生成、任务规划 |
| 引导(Steering) | 人类审批(HITL)和中断 |

额外: Harness Profile - 按 Provider/Model 打包配置为可复用 Bundle。

### 3.1 执行环境

| 工具 | 说明 |
|------|------|
| ls | 列出目录文件与元数据 |
| read_file | 读取文件(支持 offset/limit 分页) |
| write_file | 创建新文件 |
| edit_file | 精确字符串替换 |
| glob | 匹配文件模式(如 **/*.py) |
| grep | 搜索文件内容 |
| execute | 运行 Shell 命令(仅沙箱 Backend) |

### 3.2 上下文管理

| 层 | 说明 |
|------|------|
| Skills | 按需加载的领域知识(渐进式披露) |
| Memory | 持久化指令和偏好(AGENTS.md) |
| 摘要与卸载 | 自动压缩对话历史和大型工具输出 |
| Prompt Caching | 静态部分缓存(Anthropic 默认开启) |

### 3.3 委派

| 机制 | 说明 |
|------|------|
| 任务规划 | 内置 write_todos 工具跟踪多个任务 |
| 子 Agent | 临时子 Agent 在隔离上下文中执行子任务，只返回最终结果 |

### 3.4 引导(Steering)

通过 interrupt_on 参数在指定工具调用前暂停执行，等待人类审批:

```python
agent = create_deep_agent(
    model="deepseek-v4-flash",
    interrupt_on={"edit_file": True},
)
```

### 3.5 Harness Profile

```python
from deepagents import HarnessProfile, register_harness_profile
register_harness_profile(
    "anthropic:claude-sonnet-4-6",
    HarnessProfile(excluded_tools=frozenset({"ls", "read_file"})),
)
```

## 4. 与其他框架的关系

| 框架 | 定位 |
|------|------|
| LangChain | 核心组件: Chat Model、Tool、Agent 基类 |
| LangGraph | 运行时: 编排、持久化执行、流式、HITL |
| Deep Agents | 在 LangChain + LangGraph 之上提供开箱即用的 Harness |

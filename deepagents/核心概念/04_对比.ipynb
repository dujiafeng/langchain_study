# 04_对比

**来源:** [LangChain Docs — Comparison with Claude Agent SDK](https://docs.langchain.com/oss/python/deepagents/comparison)

LangChain Deep Agents 与 Claude Agent SDK 的全面对比。两者都是构建自定义 Agent 的 Harness，但在执行环境、部署方式和供应商耦合度上做出了不同取舍。Deep Agents 已在 OpenSWE 和 LangSmith Fleet 生产中使用。


## 1. 概览 (At a Glance)

| 维度 | Deep Agents | Claude Agent SDK |
|------|-------------|------------------|
| **Agent 运行位置** | 沙盒内，或沙盒外远程执行命令 | 沙盒内 |
| **执行后端** | 可插拔：本地、虚拟文件系统、远程沙盒或自定义 | 沙盒的本地文件系统 |
| **模型供应商** | 任何（Anthropic、OpenAI、Google 等 100+） | 仅 Claude（Anthropic、Bedrock、Vertex、Azure） |
| **每供应商调优** | Harness Profiles：声明式配置包 | 代码中每次调用时配置 |
| **部署** | Managed (LangSmith) 或自托管 (`langgraph build`) | 自托管（需自己构建服务器、认证和流式层） |
| **多租户** | 内置：作用域线程、每用户沙盒、RBAC | 需自行构建 |
| **许可证** | MIT | MIT（Claude Code 本身是专有） |


## 2. Agent 与执行环境

Agent 与沙盒连接有两种模式：

1. **Agent 运行在沙盒内部** — Claude Agent SDK 仅支持此模式
2. **Agent 运行在沙盒外部，将沙盒作为工具使用** — Deep Agents 支持此模式

Deep Agents 支持两种模式并提供可插拔后端：
- 在沙盒内运行 Agent（与 Claude Agent SDK 相同）
- 在长期运行的容器中运行 Agent，将远程沙盒作为工具，通过网络执行命令
- 为测试替换虚拟文件系统，或为自有基础设施使用自定义后端


## 3. 多租户 (Multi-tenancy)

生产环境中应用通常需要为每个最终用户隔离环境。

| 方面 | Deep Agents | Claude Agent SDK |
|------|-------------|------------------|
| **隔离方案** | 在 Harness 中配置每用户/每助手的沙盒、作用域线程、运行历史和 RBAC | 需自行构建 API 封装：为每个用户启动沙盒、跟踪归属、清理资源 |
| **认证代理** | 使用 LangSmith Sandbox 可开箱即用 | 需自行管理用户凭据 |


## 4. 生产级 Agent 服务器

| 方面 | Deep Agents | Claude Agent SDK |
|------|-------------|------------------|
| **服务器** | 开箱即用：流式端点、线程管理、运行历史、Webhook、认证 | 需自行构建 HTTP/WebSocket 或 SSE 服务器 |
| **部署模式** | Managed（LangSmith）或自托管（`langgraph build` → Docker 镜像） | 仅自托管 |
| **跨模型管理** | LangSmith Fleet 支持任意模型供应商 | 仅限于 Anthropic 生态系统 |


## 5. 大语言模型 (LLM)

- **Claude Agent SDK**：将模型、后端和部署捆绑，优化三者间的支持
- **Deep Agents**：模型供应商、执行后端和部署目标可独立选择，保留最大灵活性


## 6. 生态系统 (Ecosystems)

- **Claude Agent SDK**：为 Claude 和 Anthropic 产品面量身定制
- **Deep Agents**：集成更广泛的 LangChain 生态，包括 LangSmith（可观测性、评估、部署），并支持任何模型供应商


## 7. 总结 (Summary)

**选择 Deep Agents 当：**
- 需要模型和基础设施灵活性
- 需要内置多租户部署
- 希望无代码更改即可在 Managed 和自托管间切换

**选择 Claude Agent SDK 当：**
- 已投资 Anthropic 生态系统
- 希望自托管并自行构建 API、认证和多租户层

# 05_记忆与技能
**来源:** [LangChain Docs — Deep Agents Code 记忆与技能](https://docs.langchain.com/docs/deep-agents/code/memory-and-skills)


## 1. 两种定制方式

在 Deep Agents Code 中有两种定制 Agent 的方式：

- **记忆（Memory）** — `AGENTS.md` 文件和自动保存的记忆，跨会话持久化。
  用于编码风格、偏好和学到的约定。
- **技能（Skills）** — 全局和项目级别的上下文、约定、指南。
  用于执行特定任务时的上下文。

## 2. 自动记忆

Agent 在 `~/.deepagents/<agent_name>/memories/` 中自动存储信息：
```text
~/.deepagents/backend-dev/memories/
├── api-conventions.md
├── database-schema.md
└── deployment-process.md
```

工作流程：
1. **Research** — 在开始任务前搜索相关记忆
2. **Response** — 执行时不确时检查记忆
3. **Learning** — 自动保存新信息供将来使用

## 3. AGENTS.md 文件

提供始终在会话启动时加载的持久上下文：

- **全局：** `~/.deepagents/<agent_name>/AGENTS.md` — 每次会话加载
- **项目：** `.deepagents/AGENTS.md` — 从该项目中运行时加载

```markdown
# ~/.deepagents/agent/AGENTS.md（全局）
## 我的编码风格
- 使用 snake_case 命名
- 包含类型注解
- 每个函数都写 docstring
```

## 4. 技能（Skills）

技能是可复用的 Agent 能力，提供专门的工作流和领域知识。

```bash
# 创建技能
# 用户级别（~/.deepagents/<agent>/skills/）
dcode skills create test-skill

# 项目级别（.deepagents/skills/）
dcode skills create test-skill --project
```

这将生成：
```text
skills/
└── test-skill/
    └── SKILL.md
```

## 5. 技能发现路径

```text
~/.deepagents/<agent_name>/skills/
~/.agents/skills/
.deepagents/skills/
.agents/skills/
~/.claude/skills/          （实验性）
.claude/skills/            （实验性）
```

## 6. 调用技能

```bash
# 会话中直接调用
/skill:code-review
/skill:code-review review the auth module

# 启动时加载技能
dcode --skill code-review
dcode --skill code-review -m 'review the auth module'

# 列出技能
dcode skills list
dcode skills list --project
dcode skills info test-skill
```

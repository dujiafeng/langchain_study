# 04_MCP工具
**来源:** [LangChain Docs — Deep Agents Code MCP 工具](https://docs.langchain.com/docs/deep-agents/code/mcp-tools)


## 1. 什么是 MCP？

MCP（Model Context Protocol）允许你通过外部服务器扩展 Deep Agents Code 的工具集——
文件系统、API、数据库等，而无需修改 Agent 本身。

Deep Agents Code 在启动时自动发现 MCP 服务器，加载其工具，
使其与内置工具一起供 Agent 使用。

## 2. 快速开始：添加 MCP 服务器

### 创建配置文件

**用户级别（所有项目均可用）：**
```json
{
    "mcpServers": {
        "docs-langchain": {
            "type": "http",
            "url": "https://docs.langchain.com/mcp"
        }
    }
}
```

**项目级别：** 在项目根目录下放置 `.mcp.json` 文件
**项目隐藏级别：** `.deepagents/.mcp.json`

## 3. 自动发现位置

| 优先级 | 位置 | 作用域 |
|--------|------|--------|
| 1 (低) | `~/.deepagents/.mcp.json` | 用户级别，所有项目 |
| 2 | `<项目>/.deepagents/.mcp.json` | 项目级别，`.deepagents` 子目录 |
| 3 (高) | `<项目>/.mcp.json` | 项目级别，根目录（兼容 Claude Code） |

支持 `--mcp-config PATH` 添加显式配置文件（最高优先级），
以及 `--no-mcp` 禁用所有 MCP 加载。

## 4. 配置格式

### stdio 服务器（默认）
```json
{
  "mcpServers": {
    "filesystem": {
      "command": "npx",
      "args": ["-y", "@modelcontextprotocol/server-filesystem", "/tmp"],
      "env": {}
    }
  }
}
```

### SSE 和 HTTP 服务器
```json
{
  "mcpServers": {
    "remote-api": {
      "type": "sse",
      "url": "https://api.example.com/mcp",
      "headers": { "Authorization": "Bearer your-token" }
    }
  }
}
```

## 5. 工具过滤

```json
{
  "mcpServers": {
    "filesystem": {
      "command": "npx",
      "args": ["-y", "@modelcontextprotocol/server-filesystem", "/tmp"],
      "allowedTools": ["read_file", "list_directory"]
    },
    "github": {
      "command": "npx",
      "args": ["-y", "@modelcontextprotocol/server-github"],
      "env": { "GITHUB_TOKEN": "ghp_..." },
      "disabledTools": ["delete_repository"]
    }
  }
}
```

`allowedTools` 和 `disabledTools` 互斥，支持 `fnmatch` 通配符。

## 6. OAuth 登录

对于需要 OAuth 的远程 MCP 服务器：
```json
{
  "mcpServers": {
    "linear": {
      "type": "http",
      "url": "https://mcp.linear.app/mcp",
      "auth": "oauth"
    }
  }
}
```

```bash
# 运行 OAuth 登录流程
dcode mcp login linear
```

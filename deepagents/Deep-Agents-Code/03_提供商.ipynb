## 1. 支持的提供商

Deep Agents Code 支持所有 LangChain 兼容的聊天模型提供商。
以下提供商内置支持（安装对应 extras 即可）：

| 提供商 | 包名 | 凭据环境变量 | 模型配置 |
|--------|------|-------------|---------|
| OpenAI | `langchain-openai` | `OPENAI_API_KEY` | ✅ |
| Azure OpenAI | `langchain-openai` | `AZURE_OPENAI_API_KEY` | ✅ |
| Anthropic | `langchain-anthropic` | `ANTHROPIC_API_KEY` | ✅ |
| Google Gemini | `langchain-google-genai` | `GOOGLE_API_KEY` | ✅ |
| AWS Bedrock | `langchain-aws` | `AWS_ACCESS_KEY_ID` | ✅ |
| Ollama | `langchain-ollama` | (可选) | ❌ |
| Groq | `langchain-groq` | `GROQ_API_KEY` | ✅ |
| DeepSeek | `langchain-deepseek` | `DEEPSEEK_API_KEY` | ✅ |
| xAI | `langchain-xai` | `XAI_API_KEY` | ✅ |
| Perplexity | `langchain-perplexity` | `PERPLEXITY_API_KEY` | ✅ |
| OpenRouter | `langchain-openrouter` | `OPENROUTER_API_KEY` | ✅ |
| LiteLLM | `langchain-litellm` | 依提供商而定 | ❌ |

## 2. 安装提供商包

```bash
# 安装时指定 extras
DEEPAGENTS_EXTRAS="baseten,groq" curl -LsSf https://langch.in/dcode | bash

# 或使用 uv 工具安装
uv tool install 'deepagents-code[baseten,groq]'

# 后续添加包
uv tool install deepagents-code --with langchain-ollama

# 安装所有提供商
uv tool install 'deepagents-code[anthropic,baseten,bedrock,...]'
```

## 3. 切换模型

```bash
# 交互式切换
/model

# 直接指定模型
/model anthropic:claude-opus-4-7

# 启动时指定
dcode --model openai:gpt-5.5

# 设置默认模型
/model --default anthropic:claude-opus-4-7

# 清除默认模型
/model --default --clear
```

## 4. 模型解析顺序

```text
1. --model 标志（最高优先级）
2. [models].default 配置
3. [models].recent 配置（/model 自动写入）
4. 环境自动检测：OPENAI_API_KEY → ANTHROPIC_API_KEY → GOOGLE_API_KEY → GOOGLE_CLOUD_PROJECT
```

## 5. 模型参数

```python
# 通过 --model-params 传递额外参数（最高优先级，仅本次会话）
dcode --model openai:gpt-5.5 --model-params '{"reasoning": {"effort": "high"}}'

# 或通过 config.toml 持久设置
# [models.providers.anthropic.params]
# thinking = { type = "enabled", budget_tokens = 10000 }
# max_tokens = 16000
```

## 1. 什么是 Deep Agents Code？

Deep Agents Code（`dcode`）是一个基于 Deep Agents SDK 构建的开源终端编码 Agent。
它支持任意支持工具调用的 LLM，并可在会话中切换提供商或模型。

主要特性：
- **持久记忆** — 跨会话携带上下文
- **可定制技能** — 塑造 Agent 的行为
- **审批控制** — 门控代码执行
- **MCP 工具** — 通过 Model Context Protocol 加载外部工具
- **远程沙箱** — 在远程环境而非本地执行工具
- **子 Agent** — 将任务委托给专门的子 Agent

## 2. 快速开始

```bash
# 一键安装并启动
curl -LsSf https://langch.in/dcode | bash
```

添加提供商凭据后，即可向 Agent 下达任务：
```bash
# 创建一个 Python 脚本打印 "Hello, World!"
dcode
```

## 3. 内置工具一览

| 工具 | 描述 | 人工审批 |
|------|------|---------|
| `ls` | 列出文件和目录 | - |
| `read_file` | 读取文件内容 | - |
| `write_file` | 创建或覆盖文件 | 需要 |
| `edit_file` | 对现有文件进行定向编辑 | 需要 |
| `glob` | 查找匹配模式的文件 | - |
| `grep` | 跨文件搜索文本模式 | - |
| `execute` | 执行 Shell 命令 | 需要 |
| `web_search` | 使用 Tavily 搜索网页 | 需要 |
| `fetch_url` | 获取网页并转为 Markdown | 需要 |
| `task` | 将工作委派给子 Agent | 需要 |
| `ask_user` | 向用户提问 | - |
| `compact_conversation` | 压缩对话历史 | 混合 |
| `write_todos` | 创建和管理任务列表 | - |

> **注意：** 具有潜在破坏性的操作需要用户审批后方可执行。可通过 `Shift+Tab` 切换自动审批模式，或使用 `-y` / `--auto-approve` 标志启动。

## 4. 命令参考

```bash
# 使用特定 Agent 配置
dcode --agent mybot

# 使用特定模型 (provider:model 格式)
dcode --model anthropic:claude-opus-4-7

# 使用自动审批模式
dcode -y

# 非交互模式
dcode -n "Summarize this repo"
```

## 5. 启用追踪（可选）

将以下内容添加到 `~/.deepagents/.env` 或导出环境变量：
```bash
LANGSMITH_TRACING=true
LANGSMITH_API_KEY=lsv2_pt_...
LANGSMITH_PROJECT=deepagents-code
```

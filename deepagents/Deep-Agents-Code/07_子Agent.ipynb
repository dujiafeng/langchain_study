# 07_子Agent
**来源:** [LangChain Docs — Deep Agents Code 子 Agent](https://docs.langchain.com/docs/deep-agents/code/subagents)


## 1. 什么是子 Agent？

子 Agent（Subagents）允许你将专门任务委托给独立的 Agent 实例。
每个子 Agent 生活在自己的文件夹中，通过 `AGENTS.md` 文件定义：

```text
.deepagents/agents/{subagent-name}/AGENTS.md          # 项目级别
~/.deepagents/{agent}/agents/{subagent-name}/AGENTS.md # 用户级别
```

> **注意：** 目前 Deep Agents Code 仅支持同步子 Agent。异步子 Agent 暂不可用。

## 2. 文件格式

子 Agent 的 `AGENTS.md` 文件使用 YAML 前言 + Markdown 正文：

```yaml
---
name: researcher
description: Research topics on the web before writing content
model: anthropic:claude-haiku-4-5-20251001
---

You are a research assistant with access to web search.

## Your Process
1. Search for relevant information
2. Summarize findings clearly
```

| 字段 | 必需 | 描述 |
|------|------|------|
| `name` | ✅ | 子 Agent 名称 |
| `description` | ✅ | 描述，用于触发委托 |
| `model` | ❌ | 覆盖主 Agent 的模型（`provider:model` 格式） |

## 3. 优先级规则

```text
用户级别：~/.deepagents/{agent}/agents/ （较低优先级）
项目级别：.deepagents/agents/         （较高优先级）
```

项目级别的子 Agent 会覆盖同名的用户级别子 Agent。

## 4. 示例：成本高效子 Agent

为简单委托任务使用更便宜、更快的模型：

```yaml
# ~/.deepagents/agent/agents/general-purpose/AGENTS.md
---
name: general-purpose
description: General-purpose agent for research and multi-step tasks
model: anthropic:claude-haiku-4-5-20251001
---

You are a general-purpose assistant. Complete the task efficiently
and return a concise summary.
```

这将覆盖内置的通用子 Agent，将所有委托任务路由到更便宜的模型。

## 5. 使用子 Agent

```bash
# 进入 dcode 会话后，Agent 会根据任务描述自动决定是否委托
# 给子 Agent。你也可以通过 /task 工具手动触发。

# 示例：要求 Agent 研究一个主题
# "Research the latest Python 3.13 features and summarize them"
# → Agent 可能将研究部分委托给 researcher 子 Agent
```

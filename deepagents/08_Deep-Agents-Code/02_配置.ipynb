## 1. 配置文件结构

Deep Agents Code 的配置存储在 `~/.deepagents/` 目录下：

| 文件 | 格式 | 用途 |
|------|------|------|
| `config.toml` | TOML | 模型默认值、提供商设置、构造函数参数、主题等 |
| `.env` | Dotenv | 全局 API 密钥和 secrets |
| `hooks.json` | JSON | 生命周期事件订阅 |
| `.mcp.json` | JSON | MCP 服务器定义 |

## 2. 提供商凭据管理

有两种方式设置凭据：

**推荐：使用 `/auth` 命令**
```bash
# 在会话中打开凭据管理器
/auth
```

**环境变量方式（适用于 CI/CD）：**
```python
# 在 ~/.deepagents/.env 中设置
ANTHROPIC_API_KEY = "sk-ant-..."
OPENAI_API_KEY = "sk-..."

# 也可以使用 DEEPAGENTS_CODE_ 前缀隔离作用域
# DEEPAGENTS_CODE_OPENAI_API_KEY = "sk-..."
```

## 3. 环境变量加载顺序

```text
# 优先级：shell 环境 > 项目 .env > 全局 .env
# 已设置的 shell 变量不会被覆盖
```

两个 `.env` 文件在启动时被加载：
1. **项目 `.env`** — 当前工作目录下的 `.env` 文件
2. **全局 `~/.deepagents/.env`** — 所有项目的共享后备文件

## 4. 启用 Web 搜索（Tavily）

```bash
# 在 ~/.deepagents/.env 中添加
TAVILY_API_KEY = "tvly-..."

# 或使用前缀隔离
DEEPAGENTS_CODE_TAVILY_API_KEY = "tvly-..."

# 在会话中重新加载
/reload
```

## 5. config.toml 详解

```toml
# 默认和最近使用的模型
[models]
default = "ollama:qwen3:4b"
recent = "google_genai:gemini-3.5-flash"

# 提供商配置
[models.providers.ollama]
models = ["qwen3:4b", "llama3"]
base_url = "http://localhost:11434"

[models.providers.ollama.params]
temperature = 0
num_ctx = 8192

# 每个模型的参数覆盖
[models.providers.ollama.params."qwen3:4b"]
temperature = 0.5
num_ctx = 4000
```

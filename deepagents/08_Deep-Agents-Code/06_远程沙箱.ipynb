# 06_远程沙箱
**来源:** [LangChain Docs — Deep Agents Code 远程沙箱](https://docs.langchain.com/docs/deep-agents/code/remote-sandboxes)


## 1. 沙箱模式简介

Deep Agents Code 采用 **沙箱即工具** 模式：
- `dcode` 进程（LLM 循环、记忆、工具调度）在本地运行
- Agent 工具调用（`read_file`、`write_file`、`execute` 等）目标指向远程沙箱

支持的沙箱提供商：
| 提供商 | 安装命令 | 凭据环境变量 |
|--------|---------|-------------|
| LangSmith | 默认内置 | `LANGSMITH_API_KEY` |
| Daytona | `'deepagents-code[daytona]'` | `DAYTONA_API_KEY` |
| Modal | `'deepagents-code[modal]'` | Modal 设置 |
| Runloop | `'deepagents-code[runloop]'` | `RUNLOOP_API_KEY` |
| AgentCore | `'deepagents-code[agentcore]'` | AWS 凭据 |

## 2. 安装沙箱依赖

```bash
# 安装所有沙箱提供商
uv tool install 'deepagents-code[all-sandboxes]'

# 或单独安装
uv tool install 'deepagents-code[daytona]'
uv tool install 'deepagents-code[modal]'
uv tool install 'deepagents-code[runloop]'
```

## 3. 设置凭据

```bash
# LangSmith
export LANGSMITH_API_KEY="lsv2_..."

# Daytona
export DAYTONA_API_KEY="dpa_..."

# Runloop
export RUNLOOP_API_KEY="rl_..."

# AgentCore (AWS Bedrock)
export AWS_ACCESS_KEY_ID="your-key"
export AWS_SECRET_ACCESS_KEY="your-secret"
export AWS_REGION="us-west-2"
```

## 4. 使用沙箱

```bash
# 使用 LangSmith 沙箱
dcode --sandbox langsmith

# 使用 Daytona 沙箱
dcode --sandbox daytona

# 使用已存在的沙箱（跳过创建和清理）
dcode --sandbox runloop --sandbox-id dbx_abc123

# 使用设置脚本
dcode --sandbox modal --sandbox-setup ./setup.sh
```

## 5. 沙箱标志和默认工作目录

| 提供商 | 默认工作目录 |
|--------|-------------|
| LangSmith | `/root` |
| Daytona | `/home/daytona` |
| Modal | `/workspace` |
| Runloop | `/home/user` |
| AgentCore | `/tmp` |

```bash
# 示例：创建新 Daytona 沙箱
# dcode --sandbox daytona

# 示例：重用现有沙箱
# dcode --sandbox runloop --sandbox-id dbx_abc123

# 示例：创建后运行设置脚本
# dcode --sandbox modal --sandbox-setup ./setup.sh
```

## 6. 设置脚本示例

```bash
#!/bin/bash
set -e

# 克隆仓库
git clone https://github.com/username/repo.git $HOME/workspace
cd $HOME/workspace

# 持久化环境变量
cat >> ~/.bashrc <<'EOF'
export OPENAI_API_KEY="${OPENAI_API_KEY}"
cd $HOME/workspace
EOF
source ~/.bashrc
```

> **安全提示：** 沙箱隔离了代码执行，但 Agent 仍可能受到提示注入攻击。
请使用人工审批、短生命周期 secrets 和受信设置脚本。

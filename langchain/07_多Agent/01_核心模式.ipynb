# 01 - 多 Agent 核心模式

**来源:** [LangChain Docs - Multi-agent Index](https://docs.langchain.com/oss/python/langchain/multi-agent/index) | [Handoffs](https://docs.langchain.com/oss/python/langchain/multi-agent/handoffs) | [Subagents](https://docs.langchain.com/oss/python/langchain/multi-agent/subagents) | [Router](https://docs.langchain.com/oss/python/langchain/multi-agent/router) | [Skills](https://docs.langchain.com/oss/python/langchain/multi-agent/skills)

多 Agent 系统让多个专业化 Agent 协作完成复杂任务。核心模式包括：交接、子 Agent、路由和技能。

## 1. 模式对比

| 模式 | 使用场景 | 通信方式 |
|------|----------|----------|
| **交接** | Agent 之间传递对话控制权 | 节点路由 |
| **子 Agent** | 主 Agent 创建临时子任务处理器 | task 工具 |
| **路由** | 路由器 Agent 分派请求 | 条件边 |
| **技能** | Agent 按需加载专业知识 | SKILL.md 文件 |

## 2. 交接 (Handoffs)

每个 Agent 负责特定领域，遇到其他领域请求时转交。

```python
# 创建专用 Agent，通过 name 标识
weather_agent = create_agent(
    model="deepseek-v4-flash", tools=[get_weather],
    name="weather_agent",
    system_prompt="Handle weather queries.",
)
calendar_agent = create_agent(
    model="deepseek-v4-flash", tools=[check_calendar],
    name="calendar_agent",
    system_prompt="Handle scheduling.",
)
```

## 3. 子 Agent (Subagents)

主 Agent 创建临时子 Agent 处理子任务，实现上下文隔离。

```python
from deepagents import SubAgent

research_subagent: SubAgent = {
    "name": "researcher",
    "description": "Researches topics.",
    "system_prompt": "Research specialist.",
    "tools": [search_tool],
}
agent = create_agent(
    model="deepseek-v4-flash",
    middleware=[SubAgentMiddleware(subagents=[research_subagent])],
)
```

## 4. 路由器 (Router)

路由器 Agent 分析用户意图，将请求分派给对应的专用 Agent。

```python
router = create_agent(
    model="deepseek-v4-flash", tools=[], name="router",
    system_prompt=(
        "Classify user request and route to:\n"
        "- weather for weather questions\n"
        "- calendar for scheduling\n"
        "- general for other"
    ),
)
```

## 5. 技能 (Skills)

Agent 通过技能文件按需加载领域知识（渐进式披露）。

```python
agent = create_agent(
    model="deepseek-v4-flash",
    middleware=[SkillsMiddleware(backend=backend, sources=["./skills/"])],
)
```

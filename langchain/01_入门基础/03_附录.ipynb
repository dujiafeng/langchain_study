## 1. LangChain Academy

LangChain Academy 是官方的学习平台，提供全面的学习路径：

- **入门课程**：从零开始学习 LangChain
- **进阶课程**：深入 Agent、工具和记忆系统
- **实战项目**：构建真实世界的 AI 应用
- **最佳实践**：学习如何构建可靠的 Agent

## 3. 更新日志

LangChain 的完整更新日志可在 GitHub Releases 页面查看。以下为关键版本里程碑：

### v1.0.0 (2025-10-20)

- **全面重构**：所有 chains 和 agents 替换为基于 LangGraph 的单一高级 Agent 抽象
- **标准化消息内容格式**：统一跨提供商的消息格式，支持 reasoning blocks、citations、server-side tool calls
- **向后兼容**：旧版用户可安装 `langchain-classic` 包

### v0.5.3 (2026-03-15) — Deep Agents

- Deep Agents 作为基于 LangGraph 的开源 Agent 框架发布
- 提供内置规划工具、虚拟文件系统、子 Agent 生成和上下文管理

### v0.1.0 (2024-01)

- 第一个非 0.0.x 版本
- 增加了稳定性和生产级功能

### 其他重要里程碑

查看完整的版本历史，请访问 LangChain GitHub Releases 页面。

## 1. LangChain 概述

**LangChain** 提供了一个极简且高度可配置的 Agent 框架——`create_agent`。

核心公式：**Agent = Model + Harness**

- **模型 (Model)**：作为推理引擎，驱动 Agent 的决策过程
- **Harness (框架)**：围绕模型循环的一切——提示词、工具和任何塑造行为的中件

LangChain 支持所有主流模型提供商：OpenAI、Anthropic、Google、Fireworks、Ollama、AWS Bedrock、HuggingFace、OpenRouter、Azure OpenAI、Baseten 等。

```python
# 安装依赖
# pip install -qU langchain "langchain[openai]"

from langchain.agents import create_agent

# 定义一个工具函数
def get_weather(city: str) -> str:
    """获取指定城市的天气"""
    return f"{city}的天气总是晴朗！"

# 创建 Agent
agent = create_agent(
    model="openai:gpt-5.4",
    tools=[get_weather],
    system_prompt="你是一个乐于助人的助手",
)

# 调用 Agent
result = agent.invoke(
    {"messages": [{"role": "user", "content": "旧金山的天气怎么样？"}]}
)
print(result["messages"][-1].content_blocks)
```

### 框架对比

| 框架 | 定位 | 适用场景 |
|------|------|----------|
| **LangChain (`create_agent`)** | 高度可定制的 Agent 框架 | 需要精细控制 Agent 行为的场景 |
| **Deep Agents** | "开箱即用"的高级 Agent | 复杂长任务（研究、编码），内置上下文压缩、虚拟文件系统、子 Agent 生成 |
| **LangGraph** | 底层编排框架 | 需要组合确定性和 Agent 工作流的复杂场景 |
| **LangSmith** | 可观测性与评估平台 | 跟踪、调试和评估 Agent 行为，检测问题并提出修复建议 |

### 核心优势

1. **标准化模型接口**：不同提供商有独特的 API，LangChain 标准化了模型交互方式，可无缝切换提供商，避免锁定
2. **高度可配置的 Harness**：`create_agent` 是极简框架（模型、工具、提示词、循环），通过中间件扩展，每个组件处理一个关注点
3. **基于 LangGraph 构建**：利用 LangGraph 的持久化执行、人机协同、持久化等能力
4. **LangSmith 调试**：可视化工具跟踪执行路径、捕获状态转换

## 2. 设计哲学

LangChain 的核心理念：

- **LLM 是强大且具有变革性的新技术**
- **LLM 与外部数据结合时表现更佳**
- **未来的应用将越来越具有 Agent 特性**
- **构建原型容易，但构建可靠的生成 Agent 很难**

### 两大核心关注点

1. **启用最佳模型**：标准化不同提供商的模型输入/输出，使开发者可以轻松切换到最新的 SOTA 模型
2. **模型编排复杂流程**：模型不仅用于文本生成，还应编排与外部数据和计算交互的复杂流程

## 3. 组件架构

LangChain 的组件生态分为以下几个层次：

### 组件类别

| 类别 | 用途 | 关键组件 | 使用场景 |
|------|------|----------|----------|
| **模型** | AI 推理与生成 | Chat models, LLMs, Embedding models | 文本生成、推理、语义理解 |
| **工具** | 外部能力 | APIs, 数据库等 | 网页搜索、数据访问、计算操作 |
| **Agent** | 编排与推理 | ReAct agents, tool calling agents | 非确定性工作流、决策制定 |
| **记忆** | 上下文保存 | 消息历史、自定义状态 | 对话、有状态交互 |
| **检索器** | 信息访问 | 向量检索器、Web 检索器 | RAG、知识库搜索 |
| **文档处理** | 数据摄入 | 加载器、分割器、转换器 | PDF 处理、网页抓取 |
| **向量存储** | 语义搜索 | Chroma, Pinecone, FAISS | 相似性搜索、嵌入存储 |

### 核心流程

1. **输入处理**：原始数据 → 文档加载器 → 文本分割器 → 结构化文档
2. **嵌入与存储**：文档 → 嵌入模型 → 向量 → 向量存储
3. **检索**：用户查询 → 嵌入模型 → 查询向量 → 检索器 → 相关上下文
4. **生成**：聊天模型 + 工具循环 → AI 响应
5. **编排**：Agent 协调模型、工具、检索器和记忆系统

### RAG 架构

RAG（检索增强生成）流程：
- 用户问题 → 检索器获取相关文档 → 聊天模型结合上下文生成 → 信息丰富的响应

### Agent 工具循环

- 用户请求 → Agent → 是否需要工具？→ 是：调用工具 → 工具结果 → 返回 Agent 循环
- 用户请求 → Agent → 是否需要工具？→ 否：返回最终答案

### 多 Agent 系统

- 复杂任务 → 监督 Agent → 多个 Specialist Agent → 结果汇总 → 协调响应

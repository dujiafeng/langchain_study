三种消息渲染方式：Markdown 富文本、结构化输出组件、推理过程展示。

## 1. Markdown 消息

LLM 天然产生 Markdown 格式文本。渲染流水线：
1. **接收** - useStream 将流式文本累积到 msg.text
2. **解析** - Markdown 解析器转换(小于5ms for 5KB)
3. **渲染** - 解析后的输出渲染到 DOM

```python
import ReactMarkdown from "react-markdown";
import { Prism as SyntaxHighlighter } from "react-syntax-highlighter";

function Message({ msg }) {
  return (
    <ReactMarkdown
      components={{
        code({ className, children }) {
          const match = /language-(\\w+)/.exec(className || "");
          return match ? (
            <SyntaxHighlighter language={match[1]}>
              {String(children).replace(/\\n$/, "")}
            </SyntaxHighlighter>
          ) : (
            <code>{children}</code>
          );
        },
      }}
    >
      {msg.text}
    </ReactMarkdown>
  );
}
```

## 2. 结构化输出渲染

Agent 返回 Pydantic schema，前端渲染为自定义 UI 组件。

```python
function WeatherCard({ response }) {
  return (
    <div className="weather-card">
      <h3>{response.city}</h3>
      <p>{response.temperature} C - {response.condition}</p>
    </div>
  );
}

{stream.messages.map(msg =>
  msg.response ? <WeatherCard response={msg.response} /> : <Message msg={msg} />
)}
```

## 3. 推理 Token 展示

模型产生 reasoning 内容块时，可在可折叠块中展示思考过程。

```python
function ReasoningBlock({ message }) {
  const [expanded, setExpanded] = useState(false);
  if (!message.reasoning) return null;
  return (
    <div className="reasoning-block">
      <button onClick={() => setExpanded(!expanded)}>
        {expanded ? "\u9690\u85cf" : "\u663e\u793a"} \u601d\u8003\u8fc7\u7a0b
      </button>
      {expanded && <div className="reasoning-content">{message.reasoning}</div>}
    </div>
  );
}
```
